```python
print("Randuri si coloane:", dataset.shape)
print("Jucatori unici :", dataset['Name'].nunique())
```

```python
print("Top 5 nationalitati:")
print(dataset['Nationality'].value_counts().head(5))

print("\nCea mai frecventa nationalitate:")
print(dataset['Nationality'].value_counts().idxmax())
```

```python
import matplotlib.pyplot as plt

top5 = dataset['Nationality'].value_counts().head(5)

plt.figure()
plt.pie(top5, labels=top5.index, autopct='%1.1f%%')
plt.title('Top 5 Nationalitati')
plt.show()
```

```python
medii = dataset.groupby('Nationality')[['SprintSpeed', 'Acceleration']].mean()

print(medii.sort_values(by='SprintSpeed', ascending=False).head(10))
```

```python
dataset['Position'] = dataset['Position'].fillna('Unknown')
```

```python
medii_club = dataset.groupby('Club')['Overall'].mean()

print(medii_club.idxmax(), medii_club.max())
```

```python
def convert_value(x):
    x = x.replace('€', '')
    if 'M' in x:
        return float(x.replace('M', '')) * 1_000_000
    elif 'K' in x:
        return float(x.replace('K', '')) * 1_000
    else:
        return float(x)

dataset['Value_num'] = dataset['Value'].apply(convert_value)
dataset['Wage_num'] = dataset['Wage'].apply(convert_value)

rezultat = dataset[dataset['Value_num'] > dataset['Wage_num']]

print(len(rezultat))
```

```python
dataset['is_underpaid'] = dataset['Wage_num'] < (dataset['Value_num'] / 100)

print(dataset[['Name', 'Wage_num', 'Value_num', 'is_underpaid']].head(10))
```

```python
dataset["Score"] = (
    0.3 * dataset["Overall"] +
    0.3 * dataset["Potential"] +
    0.2 * dataset["SprintSpeed"] +
    0.2 * dataset["ShortPassing"]
)

display(dataset[["Name", "Score"]].head())
```

```python
df_bun = dataset[['Name', 'Wage', 'Value', 'Wage_num', 'Value_num']].copy()

df_bun['difference'] = df_bun['Value_num'] - df_bun['Wage_num']

df_bun = df_bun.sort_values(by='difference', ascending=False)

print(df_bun[['Name', 'Wage', 'Value', 'difference']].head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# scatterplot
plt.figure()
sns.scatterplot(data=dataset, x='Value_num', y='Wage_num')

plt.title('Value vs Wage')
plt.xlabel('Value')
plt.ylabel('Wage')

plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

source = ColumnDataSource(dataset)

p = figure(
    title="Value vs Wage (Hover pentru jucator)",
    x_axis_label='Value',
    y_axis_label='Wage',
    tools="pan,wheel_zoom,box_zoom,reset"
)

p.scatter(x='Value_num', y='Wage_num', size=6, source=source)

hover = HoverTool()
hover.tooltips = [
    ("Name", "@Name"),
    ("Value", "@Value"),
    ("Wage", "@Wage"),
    ("Overall", "@Overall")
]

p.add_tools(hover)

show(p)
```
